```python
import oracledb
import oml, re
import pandas as pd
from IPython.display import display
from PIL import Image
import matplotlib.pyplot as plt

def get_conn_cursor(user: str,
                    password: str,
                    dsn: str):
    conn = oracledb.connect(user=user, password=password, dsn=dsn)
    cursor = conn.cursor()
    return conn, cursor

def get_db_version(cursor):
    db_ver="""select product, version_full from product_component_version"""
    rows = cursor.execute(db_ver).fetchall()
    cols = [col[0] for col in cursor.description]
    result = pd.DataFrame(rows, columns=cols)
    return result
    
def exec_sql(cursor, qry, bind_var=None):
    """ 
    함수 사용법
    selec ... where a = :b1 and b = :b2 경우 
    bind_var = {"b1":변수, "b2":500}
    """
    #conn, cursor = get_conn_cursor(user,password,conn)
    if bind_var:
        cursor.execute(qry, bind_var)
    else:
        cursor.execute(qry)
    rows = cursor.fetchall()
    cols = [col[0] for col in cursor.description]
    result = pd.DataFrame(rows, columns=cols)
    #conn.close()
    return result

def exec_sql_rows(cursor, qry, bind_var=None):
    """ 
    함수 사용법
    selec ... where a = :b1 and b = :b2 경우 
    bind_var = {"b1":변수, "b2":500}
    """
    #conn, cursor = get_conn_cursor(user,password,conn)
    if bind_var:
        cursor.execute(qry, bind_var)
    else:
        cursor.execute(qry)
    rows = cursor.fetchall()
    #conn.close()
    return rows

def dml_sql(conn,cursor,sql):
    cursor.execute(sql)
    conn.commit()
    result = "complete commit for DML"
    #conn.close()
    return result

def ddl_sql(cursor,sql):
    cursor.execute(sql)
    result = "complete commit for DDL"
    return result

def clean_text(val):
    if isinstance(val, str):
        return re.sub(r'[^가-힣a-zA-Z0-9\s]', '', val)
    return val

def img_show(img):
    input_img = Image.open(img)
    plt.figure(figsize=(5, 4))
    plt.imshow(input_img)
    plt.axis('off')
    plt.show()
```
